```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/SalesJan2009.csv')
```

```python
df.head()
```

```python
prom_ventas_pais = df.groupby('Country')['Price'].mean().reset_index()
```

```python
pais_ciudad_sum = df.groupby(['Country', 'City'])['Price'].sum().reset_index()
```

```python
datos = {'Pais': [], 'Promedio': [], 'Ciudades': []}
for (pais, prom) in zip(prom_ventas_pais['Country'], prom_ventas_pais['Price']):
  print("Promedio ventas de ", pais, ":", prom)
  ciudades = pais_ciudad_sum[(pais_ciudad_sum['Country'] == pais) & (pais_ciudad_sum['Price'] >= prom)]['City']

  salida = []
  for ciudad in ciudades:
    salida.append(ciudad.strip())

  print("Ciudades con ventas >= prom :", ",".join(salida))
  print("----------------------------------------------")

  datos['Pais'].append(pais)
  datos['Promedio'].append(prom)
  datos['Ciudades'].append(salida)
```

```python
datos_df = pd.DataFrame(datos)
```

```python
datos_df.to_csv('salida.csv', sep=',')
```
